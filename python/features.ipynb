#### load and merge features

```python
import pandas as pd
import numpy as np
import re

base_dir = 'C:\\KinectData\\converted\\'
def readAndMergeFiles(files):
    dataFrames = []
    for file in files:
        data = pd.read_csv(base_dir+file, sep=';').drop('index', axis=1)

        p = re.compile('_\w+_(\w+)\.csv')
        names = list(map(lambda c: p.search(c).group(1), data.columns))
        data.columns = names

        dataFrames.append(data)
    return pd.concat(dataFrames, ignore_index=True, join='inner')
```

```python
features_f = readAndMergeFiles([
                               #'average_speed_eng_f_scal.csv',
                               #'distance_avg_eng_f_scal.csv',
                               'max_delta_per_frame_f_new.csv',
                               #'max_delta_with_start_position_eng_f.csv'
                               ])
features_n = readAndMergeFiles([
                               #'average_speed_rus_n_scal.csv',
                               #'distance_avg_rus_n_scal.csv',
                               'max_delta_per_frame_n_new.csv',
                               #'max_delta_with_start_position_rus_n.csv'
                               ])
```

```python
from matplotlib import pyplot as plt

def plot(values):
    plt.figure(figsize=(12,7))
    plt.plot(values,zorder=1,label='ordered by index')
    plt.plot(sorted(values),zorder=1,label='ordered by value')
```

```python
def showPointsOnPlot(points, values):
    plot(values)
    showPoints(points, values, 'r')
```

```python
def showPoints(points, values, c):
    for p in points:
        plt.scatter(p, values[p], c=c, s=20, zorder=2)
```

#### t-test

```python
from scipy.stats import ttest_ind,ttest_rel

features_t_results = []
for i in range(0,features_f.shape[0]):
    ff = features_f.values[i]
    fn = features_n.values[i]
    if 0 in ff or 0 in fn:
        ff = features_f.values[i-1]
        fn = features_n.values[i-1]
    features_t_results.append(ttest_rel(ff,fn))
```

```python
df = pd.DataFrame(features_t_results)
pLessThanAlphaFeatures = df[df.pvalue<0.05].index.values
print('count of features with p<0.05: ',pLessThanAlphaFeatures.shape[0])
```

```python
plot(list(map(lambda x: x.pvalue,features_t_results)))
plt.axhline(0.05,c='r',label='alpha-level')
plt.legend()
```

```python
plot(list(map(lambda x: abs(x.statistic),features_t_results)))
plt.legend()
```

#### separate features and labels

```python
from sklearn import preprocessing

x = all_data.drop('label', axis=1)
### normalize ###
#x = (x-x.min())/(x.max()-x.min())
x = pd.DataFrame(preprocessing.MinMaxScaler().fit_transform(x))
### --------- ###
y = all_data['label']

print(x.shape)
print(y.shape)
x.head()
```

#### fisher's score

```python
def columnN(n,tuples):
    return list(map(lambda x: x[n], tuples))
```

```python
def my_fisher(k, xs):
    scores = []
    i=-1
    for ft in xs.T.values:
        i+=1
        fn = xs.columns[i]
        
        ft_1 = ft[y == 'native']
        ft_2 = ft[y == 'foreign']
        
        p_j = 0.5

        mu = np.mean(ft)
        mu_1 = np.mean(ft_1)
        mu_2 = np.mean(ft_2)
        
        sigma_1 = np.std(ft_1)
        sigma_2 = np.std(ft_2)

        f = ( p_j*np.square(mu_1-mu) + p_j*np.square(mu_2-mu) ) / ( p_j*np.square(sigma_1) + p_j*np.square(sigma_2) )
        f = f if not np.isnan(f) else 0

        scores.append((fn,f))
    selected = sorted(scores, key=lambda t: t[1], reverse=True)[0:k]

    print('selected features: ',columnN(0, selected))
    print('with scores: ',columnN(1, selected))
    
    return { 'scores':columnN(1, scores), 'selected':selected }
```

```python
top_n = 30
```

```python
fisher_all = my_fisher(top_n, x)
idxs = columnN(0, fisher_all['selected'])

data_top_features = pd.concat([x.filter(items=x.columns[idxs]),y], axis=1)
print(data_top_features.shape)
data_top_features.head()

showPointsOnPlot(idxs, fisher_all['scores'])
plt.legend()
```

#### fisher top correlation

```python
plt.figure(figsize=(12,7))
for i in range(0,top_n-1):
    plt.plot(data_top_features.values[:,i],label=data_top_features.columns[i])
plt.legend()
```

#### most not correlated features

```python
constant_f = [24]

corr_matrix = x.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
to_drop = [column for column in upper.columns if any(upper[column] > 0.8) or all(np.isnan(upper[column]))]

x_not_correlated = x.drop(x.columns[to_drop + constant_f],axis=1)

print(x_not_correlated.shape)
x_not_correlated.corr()
```

```python
plt.figure(figsize=(12,7))
for i in range(0,len(x_not_correlated.columns)):
    plt.plot(x_not_correlated.values[:,i],label=x_not_correlated.columns[i])
plt.legend()
```

```python
fisher_noncor = my_fisher(fisher_top_n, x_not_correlated)
idxs_noncor = columnN(0, fisher_noncor['selected'])

print('p-value of features is less than alpha: ',list(map(lambda x: x in pLessThanAlphaFeatures,idxs_noncor)))
```
